```python
# prompt: How to upload a json file from an api to ms azure blob storage

import requests
import json
from azure.storage.blob import BlobServiceClient, BlobClient, ContainerClient

# Azure Blob Storage connection details
Azure_Storage_Connection_String = "DefaultEndpointsProtocol=https;AccountName=cis4400solarenergy;AccountKey=dgwaHXlBLC7hjrvbBn4HtRLcbht91uMABau3yQQY7GNPT+CnNismzuWlTalaEwHnapgntDWyCq5C+AStuE97jA==;EndpointSuffix=core.windows.net"
Container_Name = "mainstorage"

# Create BlobServiceClient
blob_service_client = BlobServiceClient.from_connection_string(Azure_Storage_Connection_String)
container_client = blob_service_client.get_container_client(Container_Name)

def upload_json_from_api(api_url, filename):
    """
    Uploads JSON data from an API to Azure Blob Storage.

    Args:
        api_url (str): The URL of the API endpoint.
        filename (str): The name of the blob file to be created in Azure Blob Storage (including .json extension).
    """
    try:
        response = requests.get(api_url)
        response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)

        data = response.json()

        # Convert JSON data to a string
        json_data_str = json.dumps(data)

        # Upload to Azure Blob Storage
        blob_client = container_client.get_blob_client(filename)
        print(f"Uploading {filename} to Azure Blob Storage...")
        blob_client.upload_blob(json_data_str, overwrite=True)
        print(f"Upload successful for {filename}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data from API: {e}")
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON response: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage (replace with your API URL and desired filename)
for i in range(1, 175, 1):
  api_url = f"https://data.ny.gov/resource/3x8r-34rs.json?$limit=50000&$offset={i}"
  filename = f"solar_energy_data_{i}.json" # Dynamic filename for each iteration
  upload_json_from_api(api_url, filename)
```
